```python
import seaborn as sns

tips = sns.load_dataset("tips")
```

```python
tips
tips.columns
```

# 1 - Visualizar la distribución de propinas en función de la factura total.

```python
sns.displot(data=tips, x="total_bill", y="tip")
```

# 2 - Igual pero destacando por la hora.

```python
sns.displot(data=tips, x="total_bill", hue="time", y="tip")
```

# 3 - Mostrar la frecuencia distinguiendo por sexo del cliente.

```python
freq_sex = tips["sex"].value_counts()
freq_sex
```

# 4 - Mostrar la frecuencia distinguiendo por sexo en función de la hora.

```python
freq_sex_time = tips.groupby("sex", observed=True)["time"].value_counts()
freq_sex_time
```

```python
sns.countplot(data=tips, x='time', hue='sex', palette="pastel")
```

# 5 - Crear el diagrama (lmplot) de la relación propina y factura en función de la hora.

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='time')
```

# 6 - Mostrar en un box-plot la factura en función de la hora.

```python
sns.catplot(data=tips, x="total_bill", hue="time", kind="box", palette="pastel")
```

# 7 - Igual que el anterior pero distinguiendo por el sexo.

```python
sns.catplot(data=tips, x="sex", y="total_bill", hue="time", kind="box", palette="pastel")
```

# 8 - Mostrar un kde de la factura distinguiendo por sexo y separando por la hora.

```python
sns.displot(data=tips, x='total_bill', hue='sex', col='time', kind='kde', fill=True, palette='Set2')
```

# 9 - Calcular el ratio y visualizar las facturas y el ratio separando en base a la hora.

```python
tips['ratio_tip_bill'] = tips['tip'] / tips['total_bill']
sns.relplot(x='total_bill', y='ratio_tip_bill', data=tips, hue="time")
```
