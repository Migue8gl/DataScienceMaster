```python
df["savings"] = pd.Series(data = [12000, 4000, 12, 45000], index = ["Miguel", "Marta", "Brian", "Elena"])
```

```python
df
```

```python
df.dropna()
```

```python
import numpy as np

df.fillna(np.mean(df.savings))
```

```python
# Se le puede aplicar una función con map a algunas columnas

def after_taxes(data):
    return data - data * 0.2

df[["savings"]].map(after_taxes)
```

```python
# Filtros
df[df["savings"] < 100]
```
